# Question 6
## Line plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('us-marriages-divorces-1867-2014.csv')
data.head()
```

```python
year=data.Year.values
marraige=data.Marriages_per_1000.values
divorces=data.Divorces_per_1000.values

plt.plot(year,marraige,color="Cyan")
plt.plot(year,divorces,color="Violet")
plt.xlabel("Number of Years")
plt.ylabel("Mariages and divoces per capita")
plt.title("The number of datas and divorcess per capita in the U.S. between 1867 and 2014. \n\n")
plt.show()
```

# Question 7
## Vertical bar chart

```python
y_1900=data.Year>=1900
y_2000=data.Year<=2000
data=data[y_1900 & y_2000]
## plt.bar(data['Marriages_per_1000'],data['Divorces_per_1000'])
plt.bar(data['Year'],data['Marriages_per_1000'],color="Grey")
plt.bar(data['Year'],data['Divorces_per_1000'],color="Black")
plt.title("Total number of Marraiges and Divorces per capita in US between 1900 to 2000 \n\n")
plt.xlabel("Between 1900 and 2000")
plt.ylabel("Marriages & Divorces per capita in US")
plt.show()
```
